```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# Load the data into a Pandas DataFrame
df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
# Inspect the first 5 rows
df.head()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = df[col].isnull().mean()
    print(f'{col} - {round(pct_missing*100,2)}%')
```

```python
# Copy only the rows with a budget and revenue greater than zero.
dfm = df.loc[(df.budget>0) & (df.revenue>0)].copy()
```

```python
# Create a new column called "year" and set it equal to the first 4 characters of the "release_date" column
dfm['year'] = dfm.release_date.str[:4]
```

```python
# Looking at unique years to search for any odd values
np.sort(dfm.year.unique())
```

```python
# Since I will be making a correlation matrix, I am curious to see if there are 
# any relationships between the genres and the other variables.
# Let's get a closer look at the unique genre combinations.
dfm.genres.unique()[:5]
```

```python
# ChatGPT came up with this code to parse the values in the genres column, create a new column for each genre,
# and mark it True/False.
import ast
# Function to extract genre names from the JSON-like data
def extract_genre_names(genre_data):
    genres = ast.literal_eval(genre_data)
    genre_names = [genre['name'] for genre in genres]
    return genre_names

# Apply the function to the 'genres' column
dfm['genre_names'] = dfm['genres'].apply(extract_genre_names)

# Create separate columns for each genre
unique_genres = set(genre for genres in dfm['genre_names'] for genre in genres)
for genre in unique_genres:
    dfm[genre] = dfm['genre_names'].apply(lambda x: genre in x)

# Drop the original 'genres' and 'genre_names' columns
dfm = dfm.drop(['genres', 'genre_names'], axis=1)
```

```python
# Let's see the results
dfm.head(3)
```

```python
# remove unnecessary and troublesome columns
dfm.drop(columns=['homepage',
                 'id',
                 'keywords',
                 'original_title',
                 'overview',
                 'production_companies',
                 'production_countries',
                 'release_date',
                 'spoken_languages',
                 'status',
                 'tagline'], 
        inplace=True)
```

```python
# Are there any duplicate rows?
any(dfm.duplicated())
```

```python
# scatter plot of revenue vs budget
sns.regplot(x="revenue", 
            y="budget", 
            data=dfm, 
            scatter_kws={"color":"red"},
            line_kws={"color":"blue"})
```

```python
# scatter plot of vote_average vs budget
sns.regplot(x="vote_average", 
            y="revenue", 
            data=dfm,
            scatter_kws={"color":"red"},
            line_kws={"color":"blue"})
```

```python
# Correlation Matrix between all numeric columns

dfm.corr(method ='pearson', numeric_only=True)
```

```python
# Let's assign a number to each text and boolean column
# First, let's find out which columns are labeled as each
dfm.dtypes
```

```python
# Now we can use the loop below to change the data type
df_num = dfm.copy()

for col in df_num.columns:
    if df_num[col].dtype == 'object':
        df_num[col] = df_num[col].astype('category').cat.codes
    elif df_num[col].dtype == 'bool':
        df_num[col] = df_num[col].astype(int)
        
```

```python
# see the results
df_num.head()
```

```python
# Let's first do a heatmap of all columns minus the genre columns

correlation_matrix = df_num.loc[:, :'year'].corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Movies")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
# Let's try again with the genres
# I want to remove title and language since they had very weak correlations

exclude_cols = ['title', 'original_language']
df_num.columns[~df_num.columns.isin(exclude_cols)]
```

```python
df_num.loc[:, ~df_num.columns.isin(exclude_cols)]
```

```python
# Correlation heatmap with genres
correlation_matrix = df_num.loc[:, ~df_num.columns.isin(exclude_cols)].corr(method='pearson')
sns.heatmap(correlation_matrix) # removing annotations because the heatmap becomes too busy
plt.title("Correlation matrix for Movies")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
# Unstack to create a single column with the correlation values
correlation_matrix = df_num.loc[:, ~df_num.columns.isin(exclude_cols)].corr(method='pearson')
corr_pairs = correlation_matrix.unstack()
print(corr_pairs)
```

```python
# sort the pairs
sorted_pairs = corr_pairs.sort_values(kind="quicksort", ascending=False)
print(sorted_pairs)
```

```python
# We can now take a look at the ones with a high correlation (> 0.5) and are not the same variable
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5].reset_index()
strong_pairs.loc[strong_pairs['level_0'] != strong_pairs['level_1']]
```
